```python
#Import packages
import pandas as pd
pd.options.mode.chained_assignment = None  # default='warn'
import datetime as dt
import math as m
import numpy as np
from scipy import stats
from matplotlib import pyplot as plt
```

```python
filename='DataLab1.xlsx'
table = pd.read_excel(filename)
table.head() # the size of the table below

table.info() # defination of the table
```

```python
#To get the date understood by python
table['Date'] = pd.to_datetime(table['Date'])
table = table.set_index('Date')

start = dt.datetime(2005, 1, 1)
end = dt.datetime(2008,12,31)
our_sample = table[start:end]
print(our_sample)
```

```python
alpha = 0.99 # VaR/ES Level
our_sample['losses'] = our_sample['PL']*-1 # Don't forget the minus sign if we want the loss distribution
sorted_losses = our_sample.sort_values('losses', ascending=False)
print(sorted_losses)
```

```python
T =  len(sorted_losses.index)# Get the number of returns
idx = m.floor((1-alpha)*T+1) # Calculate the index on the ordered data 
```

```python
print(sorted_losses.iloc[idx-1,1]) # we need iloc because we pick the row number not the index (which is a date
```

```python
print("On a 1000 USD portfolio we expect to lose less than {:.1f} USD 99% of the time".format(sorted_losses.iloc[idx-1,1]*1000))
```

```python
#time variation
rT = 500 # Size of rolling sample

our_sample.insert(2,"VaR",np.nan)

for j in range(rT,T): # Pick j from 500 up to T
    roll_sample = our_sample.iloc[j-(rT-1):j,1]
    our_sample.iloc[j,2] = np.quantile(roll_sample,0.99)
```

```python
fig,ax=plt.subplots(nrows=1,ncols=1)
ax.plot(our_sample['VaR'])
plt.xlabel('Date',fontsize=12)
plt.ylabel('VaR (%)',fontsize=12)
plt.title('HS on 500 days rolling',fontsize=16)
ax.tick_params(axis='x',labelrotation=90,labelsize=9)
```

```python
start_est = dt.datetime(2005, 1, 1)
end_est = dt.datetime(2006,12,25)
our_sample_est = table[start_est:end_est]
alpha = 0.99 # VaR/ES Level
our_sample_est['losses'] = our_sample_est['PL']*-1 # Don't forget the minus sign if we want the loss distribution
sorted_losses_est = our_sample_est.sort_values('losses', ascending=False)
print(our_sample_est)

our_sample_mu=np.mean(our_sample_est)
print (our_sample_mu)
```

```python
#2b
eta = our_sample.losses-our_sample_mu['losses'] # Setting innovation equal to Loss is ok since daily returns close to zero and hard to predict
sigma = [np.var(eta)] # Sigma_0 in slides
lambda_par = 0.94 # RiskMetrics
T = 1007
```

```python
mean_first_500 = sum(eta[:500]) / 500
variance_first_500 = sum((x - mean_first_500) ** 2 for x in eta[:500]) / 500
sigma = [lambda_par * variance_first_500]
for j in range(1,T+1):
    sigma.append((1-lambda_par)*eta[j-1]**2+lambda_par*sigma[j-1])
if "sigma" not in our_sample.columns:
    our_sample.insert(3,"sigma",np.nan)

our_sample['sigma'] = sigma
```

```python
sigma = np.array(sigma)
annualized_percent_vol = (250*sigma)**0.5
print (sigma)
```

```python
fig,ax=plt.subplots()
ax.plot(annualized_percent_vol)
plt.title('Volatility from EWMA model',fontsize=14)
```

```python
sigma_0 = [np.sqrt(sigma[0])]
for j in range(1, T +1):
    sigma_0.append(np.sqrt((1 - lambda_par) * eta[j - 1] ** 2 + lambda_par * sigma_0[j - 1]**2))

sigma_0 = np.array(sigma_0)

our_sample['sigma_0'] = sigma_0

print(our_sample)
```

```python
alpha = 0.99
window_size = 500

def rolling_large_losses(our_sample, alpha, window_size):
    large_losses = []
    for i in range(len(our_sample) - window_size + 1):
        sample_window = our_sample.losses[i:i+window_size]
        u = np.quantile(sample_window, 0.95)
        large_losses.extend(sample_window[sample_window > u])
    return large_losses

large_losses = rolling_large_losses(our_sample, alpha, window_size)
```

```python
alpha = 0.99
window_size = 500

def rolling_gpd_fit(our_sample, alpha, window_size):
    xsis = []
    locs = []
    betas = []
    for i in range(len(our_sample) - window_size + 1):
        sample_window = our_sample.losses[i:i+window_size]
        u = np.quantile(sample_window, 0.95)
        large_losses = sample_window[sample_window > u]
        if len(large_losses) > 0:
            (xsi, loc, beta) = stats.genpareto.fit(large_losses, floc=0)
            xsis.append(xsi)
            locs.append(loc)
            betas.append(beta)
    return xsis, locs, betas

xsis, locs, betas = rolling_gpd_fit(our_sample, alpha, window_size)
```

```python
alpha = 0.99
window_size = 500

def rolling_var_es(our_sample, alpha, window_size):
    var_pots = []
    es_pots = []
    for i in range(len(our_sample) - window_size + 1):
        sample_window = our_sample.losses[i:i+window_size]
        u = np.quantile(sample_window, 0.95)
        large_losses = sample_window[sample_window > u]
        if len(large_losses) > 0:
            (xsi, loc, beta) = stats.genpareto.fit(large_losses, floc=0)
            N = len(sample_window)
            N_u = len(large_losses)
            middle_part = (N/N_u)*(1-alpha)
            var_pot = u+(beta/xsi)*((middle_part**(-xsi))-1)
            es_pot = (var_pot + beta - xsi * u) / (1 - xsi)
            var_pots.append(var_pot)
            es_pots.append(es_pot)
    return var_pots, es_pots

var_pots, es_pots = rolling_var_es(our_sample, alpha, window_size)
```

```python
date_range = pd.date_range(end=pd.datetime.now(), periods=509, freq='D')
new_df = pd.DataFrame({'var_pots': np.array(var_pots)}, index=date_range)
new_df.insert(1,'es_pots',np.array(es_pots))
fig, ax = plt.subplots()
ax.plot(new_df)
plt.title('Value at risk & Expected Shortfall', fontsize=14)
ax.legend(['var_pots','es_pots'])
fig.set_size_inches(12, 6)
```
