```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import math
from scipy import stats as st
```

```python
path = "./DataLab1.xlsx"
df = pd.read_excel(path, index_col = 0)
df["losses"] = -df.PL
df.head()
```

```python
fig,ax=plt.subplots(nrows=1,ncols=1)
ax.plot(df.PL)
plt.xlabel('Date',fontsize=12)
plt.ylabel('Profit/losses',fontsize=1)
plt.title(f'Profits and losses',fontsize=12)
ax.tick_params(axis='x',labelrotation=90,labelsize=9)
```

```python
alpha = 0.99
window_length = 500
test_start_idx = 500
modelling_start_date = df.iloc[0]
test_start_date = df.iloc[test_start_idx]


VaR_index = math.floor(window_length*alpha)
VaRs = np.empty(len(df.index))
VaRs[:] = np.nan

altVaRs = np.empty(len(df.index))
altVaRs[:] = np.nan

for day in range(test_start_idx,len(df.index)):
    VaRs[day] = df.losses[day-window_length:day].sort_values(ascending = True).iloc[VaR_index]
    altVaRs[day] = np.quantile(df.losses[day-window_length:day],alpha)
    
df["VaR"] = VaRs # using the 0.99%th number in a list of ascending sorted losses
df["altVaR"] = altVaRs # using the np.quantile which linearly interpolates 
```

```python
print("Some example of VaRs:")
for i in range(0,4):
    print(20*'-')
    print(f"Date: {df.index[test_start_idx+i*100].date()}, Empirical VaR: {df.VaR[test_start_idx+i*100]}")
    print(f"Date: {df.index[test_start_idx+i*100].date()}, Int.pol emp VaR: {df.altVaR[test_start_idx+i*100]}")
```

```python
fig,ax=plt.subplots(nrows=1,ncols=1)
ax.plot(df.altVaR)
plt.xlabel('Date',fontsize=12)
plt.ylabel('VaR (thousands of Euros)',fontsize=1)
plt.title(f'BHS on {window_length} days rolling',fontsize=12)
ax.tick_params(axis='x',labelrotation=90,labelsize=9)
```

The VaR is monotonously growing throughout almost the whole period, except for the first months, which might be suspicious. However, it does not disagree with the first plot of the daily profit/losses as the losses become more and more volatile. Also the dip in risk is in aggreement with the plot of the profit/losses as an eyeball guesstimate yield a slightly more stable period 2007-01 to 2007-07

```python
#2a)
df = df[["PL", "losses", "VaR", "altVaR"]]
sample_mean = np.mean(df.losses[0:500])
df["Unexpected_losses"] = df.losses-sample_mean
```

```python
#2b)
lambda_ = 0.94
es = [0]
es[1:] = list(df["Unexpected_losses"])
es2 = np.power(es,2)
sigmas2 = [np.var(df.losses[0:500])]

for i in range(0,len(df.losses)):
    sigt = (1-lambda_)*es2[i] + lambda_*sigmas2[i]
    sigmas2.append(sigt)
    
df["sigma2"] = sigmas2[1:]
```

```python
#2c) EWMA Conditional standard deviations:
df["sigma"] = np.power(df.sigma2,0.5)
df[test_start_idx:].sample(5).sort_values(by = 'Date')
```

```python
fig,ax=plt.subplots(nrows=1,ncols=1)
ax.plot(df.sigma)
plt.xlabel('Date',fontsize=12)
plt.ylabel('Volatility estimate',fontsize=1)
plt.title(f'Estimated volatility using EWMA',fontsize=12)
ax.tick_params(axis='x',labelrotation=90,labelsize=9)
```

```python
#3 
# scale the losses using the expected volatility
# pseudo code:
#
# for each day T in test set
#     set sig_T+1 to expected volatility the day after
#     multiply all t= 0:500 previous days losses with sig_T+1/sig_t
#     save the 500 weighted losses as a column in a matrix
#
# Implementation:

nbr_rows = window_length
nbr_cols = len(df.index)-test_start_idx

vwl_matrix = np.zeros((nbr_rows,nbr_cols))

for day in range(test_start_idx, len(df.index)-1):
    weighted_losses = df.sigma[day+1]*np.divide(
        df.losses[day-window_length:day], 
        df.sigma[day-window_length:day],
    )
    vwl_matrix[:,day-test_start_idx] = weighted_losses

# using the T+1th volatility (outside test data) for the last column:
sig_last =  np.sqrt((1-lambda_)*np.power(df.Unexpected_losses[-1],2) + lambda_*df.sigma2[-1])
vwl_matrix[:,-1] = sig_last*np.divide(
        df.losses[-window_length:], 
        df.sigma[-window_length:],
    )  
```

```python
vwl_var = np.empty(len(df.index))
vwl_var[:] = np.nan

vwl_es = np.empty(len(df.index))
vwl_es[:] = np.nan

for day in range(test_start_idx,len(df.index)):
    col = day-test_start_idx    
    vwl_var[day] = np.quantile(vwl_matrix[:,col],alpha)
    
    rows = vwl_matrix[:,col] > vwl_var[day]
    
    vwl_es[day] = np.mean(vwl_matrix[rows,col])
    
df["vwl_ES"] = vwl_es
df["vwl_VaR"] = vwl_var
```

```python
plt.plot(df.vwl_ES)
plt.plot(df.vwl_VaR)
plt.legend(["ES", "VaR"])

plt.xlabel('Date',fontsize=12)
plt.ylabel('Value (thousands of Euros)',fontsize=1)
plt.title(f'VWHS on {window_length} days rolling',fontsize=12)
plt.tick_params(axis='x',labelrotation=45,labelsize=9)
```

```python
fig,ax=plt.subplots(nrows=1,ncols=1)
ax.plot(df.vwl_VaR)
ax.plot(df.VaR)
ax.plot(df.losses[test_start_idx:])
ax.legend(["VW VaR", "Basic VaR", "Losses"])

plt.xlabel('Date',fontsize=12)
plt.ylabel('Value (thousands of Euros)',fontsize=1)
plt.title(f'VWHS on {window_length} days rolling',fontsize=12)
plt.tick_params(axis='x',labelrotation=45,labelsize=9)
fig.set_figwidth(10)
fig.set_figheight(5)
```

```python
# 5a)
plt.hist(df.losses,bins = 20, density = True)
mu = np.mean(df.losses)
std = np.std(df.losses)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = st.norm.pdf(x, mu, std)
  
plt.plot(x, p, 'k', linewidth=2)
title = f"Fit Values: mean: {mu:.2f} and standard deviation: {std:.2f}"
plt.title(title)
  
plt.show()

stat, pval = st.kstest(df.losses, 'norm')
rel = "< 0.05"
if(pval > 0.05):
    rel = "> 0.05"
print(f"Kolmogorov-Smirnov test statistic: {stat:.2f}, with p-value {pval:.10f} " + rel)
```

Not very normal distribution

```python
print(f"Mean: {np.mean(df.losses):.3f}")
print(f"Standard deviation: {np.std(df.losses):.3f}")
print(f"Skewness: {st.skew(df.losses):.3f}")
print(f"Kurtosis: {st.kurtosis(df.losses):.3f}")
```

Since the skewness for the losses is positive, the skewness of the returns are negative and indicates a left-skewed distribution. It also has a high kurtosis (Scipy uses Fishers definition per default which subtracts 3, making the normal distribution have kurtosis 0) which means leptokurtic. The mean is close to zero. These three aspects make it similar to the SP500 returns. 

```python
params_df = pd.DataFrame(index = df.index)

stud = np.empty((len(params_df.index),3)) #student-t: nu, mu, sigma
stud[:] = np.nan

norm = np.empty((len(params_df.index),2)) #normal: mu, sigma
norm[:] = np.nan


st_dist = st.t
norm_dist = st.norm

for day in range(test_start_idx,len(df.index)):
    stud[day,:] = st_dist.fit(df.losses[day-window_length:day])
    norm[day,:] = norm_dist.fit(df.losses[day-window_length:day])
```

```python
params_df["t_nu"] = stud[:,0]
params_df["t_mu"] = stud[:,1]
params_df["t_sigma"] = stud[:,2]
params_df["norm_mu"] = norm[:,0]
params_df["norm_sigma"] = norm[:,1]
params_df[500:].sample(5).sort_values(by = 'Date')
```

I feel that the results are a bit odd. The mean values of norm and student differ and we get many students-t nu-parameters that are lower than 2. I might have used the fit-functions erroneously, but I can't find where.

```python
params_df["t_nu"][params_df["t_nu"] < 2] = 2.1
```

```python
para_df = pd.DataFrame(index = df.index)
para_df["t_VaR"] = st.t.ppf(alpha, params_df.t_nu, params_df.t_mu, params_df.t_sigma)
para_df["norm_VaR"] = st.norm.ppf(alpha, params_df.norm_mu, params_df.norm_sigma)
```

```python
plt.plot(para_df.t_VaR)
plt.plot(para_df.norm_VaR)
plt.legend(["t-dist.", "normal dist."])

plt.xlabel('Date',fontsize=12)
plt.ylabel('VaR (thousands of Euros)',fontsize=1)
plt.title(f'Parametric on {window_length} days rolling',fontsize=12)
plt.tick_params(axis='x',labelrotation=45,labelsize=9)
```

```python
perc = 0.95

pot_var = np.empty(len(df.index))
pot_var[:] = np.nan
pot_es = np.empty(len(df.index))
pot_es[:] = np.nan

for day in range(test_start_idx,len(df.index)):
    current_window = df.losses[day-window_length:day]
    u = np.quantile(current_window, perc)
    large_losses = current_window[current_window > u]
    (xi, loc, beta) = st.genpareto.fit(large_losses, floc=0)
    N = window_length
    N_u = len(large_losses) # should be always be 0.05*500?
    
    pot_var[day] = u + beta/xi*(np.power(N/N_u*(1-alpha),-xi)-1)
    pot_es[day] = (pot_var[day]+beta-xi*u)/(1-xi)
```

```python
para_df["pot_VaR"] = pot_var
para_df["pot_ES"] = pot_es
```

```python
plt.plot(para_df.pot_VaR)
plt.plot(para_df.pot_ES)
plt.legend(["POT VaR", "POT ES"])

plt.xlabel('Date',fontsize=12)
plt.ylabel('VaR (thousands of Euros)',fontsize=1)
plt.title(f'Parametric on {window_length} days rolling',fontsize=12)
plt.tick_params(axis='x',labelrotation=45,labelsize=9)
```
