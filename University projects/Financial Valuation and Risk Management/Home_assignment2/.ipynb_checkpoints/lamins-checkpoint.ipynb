```python
#Christoffersen independence test BHS 2007

n1BHS = sum(our_sample2007['violations_BHS'])
n0BHS = len(our_sample2007) - n1BHS

n11BHS = 0
n00BHS = 0
n01BHS = 0
n10BHS = 0

for r in range(0,len(our_sample2007)-1):
    if our_sample2007['violations_BHS'][r]==1 and our_sample2007['violations_BHS'][r+1]==1:
        n11BHS += 1
    if our_sample2007['violations_BHS'][r]==0 and our_sample2007['violations_BHS'][r+1]==0:
        n00BHS += 1
    if our_sample2007['violations_BHS'][r]==0 and our_sample2007['violations_BHS'][r+1]==1:
        n01BHS += 1
    if our_sample2007['violations_BHS'][r]==1 and our_sample2007['violations_BHS'][r+1]==0:    
        n10BHS += 1
# Empirical estimates of all needed probabilities
pi00BHS = n00BHS / (n00BHS+n01BHS) # Slide 8 VL 9
pi01BHS = n01BHS / (n00BHS+n01BHS)
pi10BHS = n10BHS / (n10BHS+n11BHS)
pi11BHS = n11BHS / (n10BHS+n11BHS)
pi0BHS = n0BHS / (n1BHS+n0BHS)
pi1BHS = n1BHS / (n1BHS+n0BHS)         
#Check so that pi00BHS+pi01BHS and pi10BHS+pi11BHS both add to one
print("Check one: {:.3f}".format(pi00BHS+pi01BHS))
print("Check two: {:.3f}".format(pi10BHS+pi11BHS))    
lnNull = np.log(pi0BHS**n0BHS*pi1BHS**n1BHS) # Slide 9 VL 9, second equation 
lnAlt = np.log(pi00BHS**n00BHS*pi01BHS**n01BHS*pi10BHS**n10BHS*pi11BHS**n11BHS) # Slide 9 VL 9, first equation
LRindBHS = -2*(lnNull-lnAlt)
print("The LR_ind test statistic is {:.4f}".format(-2*(lnNull-lnAlt)))
# Pvalue for test stat that is chisquare(1) distributed (upper since right tail for one-sided test)
pValBHS = 1-stats.chi2.cdf(LRindBHS,1) # 1 minus since right tail for one-sided test
print("We get a p-value of {:.4f}."\
      .format(pValBHS))
```
