```python
import pandas as pd
import numpy as np
import datetime as dt
import math as m
from scipy import stats
from scipy.stats import norm
from matplotlib import pyplot as plt
from scipy.optimize import minimize
from matplotlib import cm
```

```python
def my_merton(x,sige,E,d,rf,t):
     d1 = (np.log(x[1]/d)+(rf+0.5*x[0]**2)*t)/(x[0]*np.sqrt(t))
     d2 = d1-x[0]*np.sqrt(t)
     f1 = x[1]*norm.cdf(d1)-np.exp(-rf*t)*d*norm.cdf(d2)-E # This is the equation on slide 21 Video L13 
     f2 = (x[1]/E)*norm.cdf(d1)*x[0]-sige # This is the equation on slide 22 Video L13 
     return np.sqrt(f1**2+f2**2) # Minimize f over x(1) and x(2) solves for Equity vol and Asset value
    
Accounting_data = pd.read_excel('DataLab3accounting.xlsx')
Price_data = pd.read_excel('DataLab3prices.xlsx')
Risk_free = pd.read_excel('DataLab3rf.xlsx')

Price_data['Date'] = pd.to_datetime(Price_data['Date'], format='%Y%m%d')
Price_data_new = Price_data.set_index('Date')
rf = Risk_free['Riskfree'].mean()

equityReturns = Price_data['Prices'].pct_change().dropna()
sige = stats.tstd(equityReturns)*np.sqrt(250) 
E = Accounting_data.CSHO[0]*Price_data.Prices.iloc[-1] 
ks = np.arange(0.1, 1, 0.1)
rf = np.mean(Risk_free.Riskfree) # Risk free rate as average of monthly rates
t = 1 # We want PD in 1 year
x0 = (0.5*sige,Accounting_data.LCT[0]+E)
PD = []

for k in ks: 
    K = Accounting_data.LCT + k * Accounting_data.DLTT
    res = minimize(my_merton, x0, method='BFGS',
                   args=(sige,E,K[0],rf,t), options={'disp': False})
    DD = (np.log(res.x[1])+(rf-0.5*res.x[0]**2)*np.sqrt(t)-np.log(K))/(res.x[0]*np.sqrt(t)) # DD slide 9 VL 13
    PD.append(float(norm.cdf(-DD)))

plt.plot(ks, PD)
plt.title('Sensitivity when incorporating long-term debt')
plt.xlabel('Value of k')
plt.ylabel('Value of Probability of Default')
```

```python
nmonths = [1,3,6,9,12]
ks = np.arange(0.1,1,0.1)
pd_surf = np.zeros((len(ks),len(nmonths)))


for i in range(0,len(nmonths)):
    sig = sige[i]
    c = 0
    for j in ks:
        debt = Accounting_data.LCT +j*Accounting_data.DLTT
        x0 = (sig, debt[0]+E[0])
        res = minimize(my_merton, x0, method='BFGS',args=(sig,E[0],debt[0],rf,t), options={'disp':False})
        dd = (np.log(res.x[1])+(rf-0.5*res.x[0]**2)*np.sqrt(t)-np.log(debt))/(res.x[0]*np.sqrt(t))
        pd_surf[c,i] = float(norm.cdf(-dd))
        c += 1

X, Y = np.meshgrid([1,3,6,9,12],np.arange(0.1,1,0.1))
fig = plt.figure(figsize=(12,10))
ax = fig.add_subplot(111, projection ='3d')
surface = ax.plot_surface(X,Y,pd_surf,cmap=cm.coolwarm)
ax.set_xlabel('# months for vol est')
ax.set_ylabel('k')
ax.set_zlabel('Probabilty of Default (PD)')
ax.set_title('PD as function of estimaton horizon and weight on LTD')
fig.colorbar(surface,shrink = 0.5, aspect = 8)
plt.show()
```
